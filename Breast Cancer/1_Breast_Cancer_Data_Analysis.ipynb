```python
gb = breastcancer.groupby('diagnosis')    
B_df = [gb.get_group(x) for x in gb.groups][0] #makes a list with 1st element B and seconf element as M. assigning the first element to B and the seond one to M
M_df = [gb.get_group(x) for x in gb.groups][1]
```

```python
from matplotlib import pyplot as plt

#radius_mean 
#texture_mean
#perimeter_mean 
#area_mean
#smoothness_mean
#compactness_mean
#concavity_mean
#concave_points_mean
#symmetry_mean
#fractal_dimension_mean

#x = [B_df['radius_mean']]
#y = [random.gauss(4,2) for _ in range(400)]

plt.hist(B_df['radius_mean'],bins=7, label='Benign', color ='darkslategrey')
plt.hist(M_df['radius_mean'],bins=7, label='Malignant', color = 'lightsteelblue', alpha = 0.8)
plt.legend(loc='upper right')
plt.title('Mean Radius', size = 20)
#plt.ylabel('')
#plt.xlabel('some numbers')
plt.show()
```
